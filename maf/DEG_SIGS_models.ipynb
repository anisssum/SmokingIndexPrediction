```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from itertools import combinations
from scipy.stats import pearsonr
import lightgbm as lgb
import shap
```

### Data preparation

```python
data = pd.read_csv("../data/X_tpm_DF.csv", sep = ',')
df_id = pd.read_csv("../data/tcga.tcga.LUAD.MD", sep = '\t')
```

```python
df_id.head()
replace_dict = dict(zip(df_id['external_id'], df_id['tcga_barcode']))
data['Unnamed: 0'] = data['Unnamed: 0'].replace(replace_dict)
```

```python
data
```

```python
def process_id(id):
    parts = id.split('-')
    if len(parts) > 4:
        return '-'.join(parts[:4])
    else:
        return id

# Применение функции к столбцу
data['Unnamed: 0'] = data['Unnamed: 0'].apply(process_id)
data
```

```python
data = data.set_index('Unnamed: 0')
```

```python
data
```

```python
X = pd.read_csv("./data/combined_table_signatures_tobacco.tsv", sep = '\t', index_col=1)
```

```python
X.head()
```

```python
merge_data = data.join(X)
```

```python
merge_data
merge_data = merge_data.dropna()
```

### Create models

```python
X = merge_data.drop(columns=['tobacco_smoking_history'])
y = merge_data['tobacco_smoking_history']
y = y.replace([1, 2, 3, 4, 5], [1, 2, 1, 2, 2])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)
print(f1_score(y_test, y_pred, average='macro'), f1_score(y_test, y_pred, average='micro'), f1_score(y_test, y_pred, average='weighted'))
```

```python
random_tree = RandomForestClassifier()
random_tree.fit(X_train, y_train.values.ravel())
y_pred = random_tree.predict(X_test)
print(f1_score(y_test, y_pred, average='macro'), f1_score(y_test, y_pred, average='micro'), f1_score(y_test, y_pred, average='weighted'))
```

```python
lgbm = LGBMClassifier()
lgbm.fit(X_train, y_train.values.ravel())
y_pred = lgbm.predict(X_test)
print(f1_score(y_test, y_pred, average='macro'), f1_score(y_test, y_pred, average='micro'), f1_score(y_test, y_pred, average='weighted'))
```

```python
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': [None, 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

```python
CV_rfc = GridSearchCV(estimator=random_tree, param_grid=param_grid, cv= 5)
CV_rfc.fit(X_train, y_train.values.ravel())
CV_rfc.best_params_
```

```python
rfc1=RandomForestClassifier(max_features='log2', n_estimators= 200, max_depth=5, criterion='gini')
rfc1.fit(X_train, y_train.values.ravel())
pred=rfc1.predict(X_test)
print(f1_score(y_test, pred, average='macro'), f1_score(y_test, pred, average='micro'), f1_score(y_test, pred, average='weighted'))
```

```python
rfc1=RandomForestClassifier(n_estimators= 500, max_depth=5, criterion='entropy')
rfc1.fit(X_train, y_train.values.ravel())
pred=rfc1.predict(X_test)
print(f1_score(y_test, pred, average='macro'), f1_score(y_test, pred, average='micro'), f1_score(y_test, pred, average='weighted'))
```

```python
categoricals = X_train.columns.values.tolist()
categoricals = [0, 1,2,3,4]
params = {
    "early_stopping_rounds": 50,
    "verbose_eval": 1000
}
lgb_train = lgb.Dataset(X_train,y_train,categorical_feature=categoricals)
lgb_test = lgb.Dataset(X_test,y_test,categorical_feature=categoricals)
```

```python
model = lgb.train(params, lgb_train, 10000, valid_sets=[lgb_test])
```

```python
y_pred = model.predict(X_test, num_iteration=model.best_iteration)
y_pred_binary = (y_pred < 1.5).astype(int)
y_pred_binary = [2 if x > 1.55 else 1 for x in y_pred]
print(f1_score(y_test, y_pred_binary, average='macro'), f1_score(y_test, y_pred_binary, average='micro'), f1_score(y_test, y_pred_binary, average='weighted'))
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)
```
